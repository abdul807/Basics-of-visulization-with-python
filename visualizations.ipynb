```python
sns.rugplot(data,height = 0.3)
x_min = data.min()-2
x_max = data.max()+2

x_axis = np.linspace(x_min,x_max,100)

bandwith = ((4*data.std()**5)/ (3*len(data))) ** 0.2  

kernel_list = []

for data_point in data:
    kernel = stats.norm(data_point,bandwith).pdf(x_axis)
    kernel_list.append(kernel)
    
    
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis,kernel,color='gray')
    
```

```python
sum_of_kde = np.sum(kernel_list,axis=0)
fig = plt.plot(x_axis,sum_of_kde,color='indianred')
sns.rugplot(data)
plt.yticks([])
plt.suptitle('sum of the basis of functions')
```

```python
flight_dframe = sns.load_dataset('flights')
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot_table( values='passengers',
    index='month',
    columns='year',
    aggfunc='mean',)
```

```python
flight_dframe
```

```python
flight_drame = flight_dframe.pivot(index='month', columns='year', values='passengers')
```

```python
flight_drame
```

```python
sns.heatmap(flight_drame)
```

```python
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_drame,annot = True,fmt='d')
```

```python
sns.heatmap(flight_drame,annot = True,fmt='d',
            center=flight_drame.loc['Jan',1955])
```

```python
f,(axis1,axis2)= plt.subplots(2,1)

yearly_flights = flight_drame.sum()
years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)
flights = pd.Series(yearly_flights.values)
flights =  pd.DataFrame(flights)
year_dframe = pd.concat((years,flights),axis = 1)
year_dframe.columns=['Years','flights']
# year_dframe
sns.barplot('Years',y='flights',data=year_dframe,ax=axis1)
sns.heatmap(year_dframe,annot = True,fmt='d',
            cmap='Blues',ax=axis2,
           cbar_kws ={'orientation':'vertical'})
```
